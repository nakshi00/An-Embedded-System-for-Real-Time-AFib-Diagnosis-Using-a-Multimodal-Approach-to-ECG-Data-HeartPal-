```python
from PIL import Image

def get_image_size(image_path):
    try:
        with Image.open(image_path) as img:
            width, height = img.size
            return width, height
    except IOError:
        print("Unable to open image file:", image_path)

# Example usage:
image_path_HF = "D:\CSE499\ECG Data HF\Abnormal Heartbeat Patients\HB(1).jpg"
width, height = get_image_size(image_path_HF)
print("Width:", width)
print("Height:", height)
```

```python
from PIL import Image, ImageFilter
import os

def preprocess_images(input_folder, output_folder, top, bottom, left, right, target_size=(224, 224), blur_radius=2):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all files in the input folder
    files = os.listdir(input_folder)

    for i, file_name in enumerate(files):
        try:
            # Open the image
            with Image.open(os.path.join(input_folder, file_name)) as img:
                # Crop the image
                cropped_img = img.crop((left, top, img.width - right, img.height - bottom))

                # Apply Gaussian blurring
                blurred_img = cropped_img.filter(ImageFilter.GaussianBlur(blur_radius))

                # Resize the image
                resized_img = blurred_img.resize(target_size)

                # Decrease resolution
                resized_img.thumbnail(target_size)

                # Save the preprocessed image to the output folder with an index number as the filename
                output_path = os.path.join(output_folder, f"{i}.jpg")
                resized_img.save(output_path)
                print(f"Preprocessed {file_name} successfully. Saved as {output_path}")
        except Exception as e:
            print(f"Error processing {file_name}: {e}")

# Define input and output folders
input_folder = r"D:\CSE499\ECG Data HF\Myocardial Infarction Patients"
output_folder = r"D:\CSE499\ECG Data HF\Preprocessed_MI_HB"

# Define preprocessing parameters (top, bottom, left, right, target_size, blur_radius)
top = 350
bottom = 100
left = 100
right = 100
target_size = (224, 224)
blur_radius = 2

# Preprocess images in the input folder and save them to the output folder
preprocess_images(input_folder, output_folder, top, bottom, left, right, target_size, blur_radius)
```

```python
from PIL import Image
import os

def rotate_and_save_images(input_folder, output_folder, rotation_angle=1.2):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all files in the input folder
    files = os.listdir(input_folder)

    for i, file_name in enumerate(files):
        try:
            # Open the image
            with Image.open(os.path.join(input_folder, file_name)) as img:
                # Get image dimensions
                width, height = img.size

                # Rotate the image by 1.7 degrees from each corner
                for j, angle in enumerate([rotation_angle, -rotation_angle]):
                    for k, (x, y) in enumerate([(0, 0), (width, 0), (0, height), (width, height)]):
                        rotated_corner = img.rotate(angle, resample=Image.BICUBIC, center=(x, y))
                        rotated_corner.save(os.path.join(output_folder, f"{i}_corner_{j}_{k}.jpg"))

                print(f"Augmented {file_name} successfully.")
        except Exception as e:
            print(f"Error processing {file_name}: {e}")


# Define input and output folders
input_folder = r"D:\CSE499\ECG Dataset\ECG_Mixer\Normal"
output_folder = r"D:\CSE499\ECG Dataset\ECG_Mixer\Augmented\Normal"

# Perform rotational data augmentation on images in the input folder and save them to the output folder
rotate_and_save_images(input_folder, output_folder)
```
