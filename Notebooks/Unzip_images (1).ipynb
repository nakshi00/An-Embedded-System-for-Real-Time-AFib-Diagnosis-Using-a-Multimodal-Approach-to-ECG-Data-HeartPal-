```python
def unzip_folder(zip_path, output_dir):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        # Extract all contents into the specified directory
        zip_ref.extractall(output_dir)

# Example usage:
zip_file_path = '/content/gdrive/MyDrive/ECG_IMG_all/MI.zip'  # Replace with the path to your zip file in Google Drive
output_directory = '/content/gdrive/MyDrive/ECG_IMG_all/ECG_IMG1'  # Replace with the directory where you want to extract the contents
unzip_folder(zip_file_path, output_directory)
```
